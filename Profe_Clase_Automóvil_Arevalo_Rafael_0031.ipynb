```python
# Clase automovil, atributos, marca, placas, precio
# funciones leeerdatos(), mostrardatos()
# Crear 3 objetos carro1, carro2, carro3

class Automovil:
  def __init__(self):
    self.marca = None # Initialize attributes
    self.placa = None
    self.precio = None

  def leerdatos(self):
    self.marca = input("Ingrese la marca: ")
    self.placa = input("Ingrese la placa: ")
    self.precio = input("Ingrese el precio: ")
    # These methods modify self, so returning them might not be necessary
    # if the goal is just to set attributes and print.
    # However, keeping them for consistency with original code's return statement.
    return self.marca, self.placa, self.precio

  def mostrardatos(self):
    print("Marca: ", self.marca)
    print("Placa: ", self.placa)
    print("Precio: ", self.precio)
    return self.marca, self.placa, self.precio
# Fin de la clase automovil

# Creando los objetos carro1, carro2, carro3
# Correctly instantiate objects without arguments, as __init__ doesn't take them
carro1 = Automovil()
carro2 = Automovil()
carro3 = Automovil()

print("Carro Numero 1")
carro1.leerdatos()
carro1.mostrardatos()
print("-*-*-*-*-*-*-*-*-*-*-*-*")
print("Carro Numero 2")
carro2.leerdatos()
carro2.mostrardatos()
print("-*-*-*-*-*-*-*-*-*-*-*-*")
print("Carro Numero 3")
carro3.leerdatos()
carro3.mostrardatos()
print("-*-*-*-*-*-*-*-*-*-*-*-*")

print("Programa realizado por Arevalo Rafael NL=3")
```
